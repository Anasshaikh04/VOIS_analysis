```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Profit vs. Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Sales', data=df, hue='Category')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.title('Discount Impact on Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Discount', y='Profit', data=df, hue='Category')
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount Impact on Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Discount'], bins=20, kde=True)
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.title('Distribution of Discounts')
plt.show()
```
